```python
## BlackJack Game, Python Milestone Project 2.
'''Simple player vs Computer(dealer) BlackJack game.'''

##Dictionary and List of Suit,Value and Rank
values={"Two":2,"Three":3,"Four":4,"Five":5,"Six":6,"Seven":7,"Eight":8,"Nine":9,"Ten":10,
       "Jack":10,"Queen":10,"King":10,"Ace":11}
suits=["Club","Spade","Diamond","Heart"]
ranks=["Two","Three","Four","Five","Six","Seven","Eight","Nine","Ten","Jack","Queen","King","Ace"]
```

```python
class Card:
    '''
    Takes the card Rank and Suit attribute as argument both with first letter capital
    Assigns a value attribute depending on Rank
    '''
    def __init__ (self,rank,suit):
        '''Instantiates a Card object with Rank, Suit and Value as attributes'''
        self.rank=rank
        self.suit=suit
        self.value=values[rank]
        
    def __str__(self):
        '''Returns the Rank and Suit of Card object as text'''
        return f"{self.rank} of {self.suit}"
        
```

```python
class Deck:
    '''Creates the deck and handles card dealings in and out'''
    
    def __init__(self):
        '''Instantiates a deck by creating cards using nested for loop'''
        self.deck=[Card(rank,suit) for suit in suits for rank in ranks]
        
    def shuffle(self):
        '''This functions Shuffles the card in the deck'''
        from random import shuffle
        shuffle(self.deck)
    
    def draw_card(self):
        '''Draw one card from the deck'''
        return self.deck.pop(0)
        
            
    def __str__(self):
        '''prints every card in a deck'''
        string="-- Deck --"
        for cards in self.deck:
            string+="\n"+cards.__str__()
        return string
```

```python
class Player:
    '''Creates a players instance which manages name, balance and the handattributes'''
    def __init__ (self,name,balance):
        self.name=name
        self.balance=balance
        self.hand=[]
    
    def withdraw(self,amount):
        '''a function to withdraw the ammount from player'''
        self.balance-=amount
        print(f"{self.name} now has {self.balance} coins in balance")
        
    def deposit(self,amount):
        '''a function to deposit the ammount to player'''
        self.balance+=amount
        print(f"{self.name} now has {self.balance} coins in balance")
            
    def findvalue(self):
        '''finds the value of a list of card objects'''
        value=0
        for card in self.hand:
            value+=card.value
        
        # changes ace value to 1 if value exceeds 21
        if value>21:
            for card in self.hand:
                if card.rank=="Ace":
                    value-=10
                    card.value=1
                    card.rank="`Ace"
                    break
        return(value)

    def bustcheck(self):
        '''check if value is >21 thereby busted'''
        if self.findvalue()>21:
            return True
        else:
            return False
        
    def printcards(self):
        '''Prints the detail and value of cards in hand'''
        print(f"{self.name} Cards \n-------------------------------------------------------------------")
        for i in self.hand:
            print(i,end='')
            print(f"({i.value}) ,",end=' ')
        print("Hand Value:",self.findvalue())
        print("\n")

    
    def __str__(self):
        '''Prints the player name and balance'''
        return f"{self.name} currently has {self.balance} coins in balance"
```

```python
##Game Logic 1

## create players

Name=input("Please enter your name: ")

#validate that balance is an integer
while True:
    try:
        Balance=int(input("Please enter the amount you wish to play with (in digits)"))
        break
    except:
        print("That was not a proper number \nPlease try again and enter in digits")

player=Player(Name,Balance)
computer=Player("Computer",9999999999)
print(player)
```

```python
# Game Logic 2

roundcount=0

if input("Do you want to play a round (y/n)")=="y":
    
    # while loop to keep rounds repeating
    #Create deck and player
    deck=Deck()
    deck.shuffle()
    while True:
        roundcount+=1
        print("Round:",roundcount)

        # take bet and validate it is an integer
        while True:
            try:
                bet=int(input("Please enter your bet (in digits)"))
                if bet<=player.balance:
                    player.withdraw(bet)
                    break
                else:
                    print("You don't have that much in balance, Please enter according to your balance.")
                    print(player)
                    continue
            except:
                print("That was not a proper number \nPlease try again and enter in digits")

        # initiate two empty list representing player hand
        deck.shuffle()
        player.hand=[]
        computer.hand=[]
        #place two cards in both hands
        for i in range(2):
            player.hand.append(deck.draw_card())
            computer.hand.append(deck.draw_card())

        # Display player and Computer cards
        player.printcards()
        print(f"Computer Cards\n - - - - - - - - -- - - - - -  -- - - - -")
        print(computer.hand[0],end='')
        print(f"({computer.hand[0].value}) , **hidden** ")


        # Check for blackjack
        if player.findvalue()==21:
            print("Player Wins with BlackJack")
            print(f"Player wins {bet*1.5} coins (1.5x bet)")
            player.deposit(bet*1.5)

        # Ask if player wants to hit
        while not player.bustcheck():
            if input("Do you want to hit? (y/n)")=='y':
                player.hand.append(deck.draw_card())
                player.printcards()
            else:
                break
                
        # BlackJack check        
        if player.findvalue()==21:
            computer.printcards()
            print("Player Wins with BlackJack")
            print(f"Player wins {bet*1.5} coins (1.5x bet)")
            player.deposit(bet*1.5)
            if input("Do you want to continue? (y/n)")=='y':
                continue
            else:
                break

        # bustcheck player
        elif player.bustcheck():
            computer.printcards()
            print(f"Player Loses!! player got busted with {player.findvalue()}pts in hand, Computer wins with {computer.findvalue()}pts in hand")
            print(f"Player loses {bet} coins")
            print(player)
            if input("Do you want to continue? (y/n)")=='y':
                continue
            else:
                break
            
        #Reveal Computer Cards
        computer.printcards()

        # if less than 17 hit otherwise stay
        while computer.findvalue()<17:
            if not computer.bustcheck():
                computer.hand.append(deck.draw_card())
                computer.printcards()

        #Evaluate Computer Cards and determine winner

        # bust check computer
        if computer.bustcheck():
            print(f"Player Wins!! with {player.findvalue()}pts in hand, Computer got Busted with {computer.findvalue()}pts in hand")
            print(f"Player wins {bet*2} coins (2x bet)")
            player.deposit(bet*2)

        # blackjack check computer
        elif computer.findvalue()==21:
            print("Computer Wins with BlackJack!!")
            print(f"Player loses {bet} coins")
            print(player)

        # draw check
        elif player.findvalue()==computer.findvalue():
            print(f"Draw!! Player has {player.findvalue()}pts in hand, Computer has {computer.findvalue()}pts in hand")
            print(f"Player gets their {bet} coins back")
            player.deposit(bet)

        #player win check
        elif player.findvalue()>computer.findvalue():
            print(f"Player Wins!! with {player.findvalue()}pts in hand which is higher than Computer's {computer.findvalue()}pts")
            print(f"Player wins {bet*2}coins (2x bet)")
            player.deposit(bet*2)

        #player lost
        else:
            print(f"Player loses with {player.findvalue()}pts in hand which is less than Computer's {computer.findvalue()}pts")
            print(f"Player loses {bet}coins")
            print(player)

        # Add the cards in hand back into the deck
        player.hand.clear()
        computer.hand.clear()
        # ask to continue to next round
        if input("Do you want to continue? (y/n)")=='y':
            continue
        else:
            break      

```
